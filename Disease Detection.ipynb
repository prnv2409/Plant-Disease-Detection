```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import tensorflow as tf
from keras.models import Sequential, load_model
from keras.layers import Dense, Activation, Conv2D, MaxPooling2D, Flatten, Dropout, BatchNormalization
#from keras.optimizers import SGD
from tensorflow.keras.optimizers import SGD
```

```python
from os import listdir
```

```python
directory_root = "/home/ankkitsharma/Desktop/freelancer/DeepLearning/Plant/train"
print(len(listdir(directory_root)))
```

```python
train_data = tf.keras.utils.image_dataset_from_directory(
    directory_root,
    image_size=(72, 72),
    shuffle=True,
    seed=None,
    validation_split=None,
    label_mode='categorical'
)
```

```python
validation_Data_dir ="New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/valid"
```

```python
validation_Data = tf.keras.utils.image_dataset_from_directory(
    validation_Data_dir,
    image_size=(72, 72),
    shuffle=True,
    seed=None,
    validation_split=None,
    label_mode='categorical'
)
```

```python
input_shape = [72, 72]


model = tf.keras.applications.Xception(
    input_shape=input_shape+[3],weights='imagenet',include_top=False
)

model.summary()
```

Start 

```python
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.models import Model

x=Flatten()(model.output)
predictions=Dense(38,activation='softmax')(x)

model1=Model(inputs=model.input,outputs=predictions)
```

```python
model1.summary()
```

```python
loss =tf.keras.losses.CategoricalCrossentropy()
```

```python
model1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model1.fit(train_data, epochs=3, verbose=1, validation_data=validation_Data)
```

```python
 model1.save('Xception')
```
